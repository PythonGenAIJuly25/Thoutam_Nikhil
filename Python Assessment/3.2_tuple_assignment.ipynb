Assignment 15: Tuple Functions
Write functions that take a tuple and return the minimum, maximum, and sum of the elements. Print the results for a sample tuple.

```python
def max(tuple1):
  max=tuple1[0]
  for i in tuple1:
    if i>max:
      max=i
  return max
def min(tuple1):
  min=tuple1[0]
  for i in tuple1:
    if i<min:
      min =i
  return min
def sum(tuple1):
  sum=0
  for i in tuple1:
    sum+=i
  return sum

tuple1=(34,323,3,3,43,4,4)
print(max(tuple1))
print(min(tuple1))
print(sum(tuple1))
```
